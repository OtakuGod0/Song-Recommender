```python
df = pd.read_csv('combined.csv')
```

```python
key_words = ['acoustic', 'remix', 'live', 'version', 'edit', 'demo', 'mix', 'dub', 'solo']
```

```python
df['song_formatted'] = df['song'].apply(lambda x: x.lower().replace("'", ""))
```

```python
df.drop_duplicates(['song_formatted', 'artist'], inplace = True)
```

```python
drop_index = []

# Iterate over each row by index and value
for idx, song_formatted in df['song_formatted'].items():
    # Check if any keyword is in the current song formatted string
    if any(key_word in song_formatted for key_word in key_words):
        drop_index.append(idx)  # Append the index if the condition is met
```

```python
df.drop(index = drop_index, inplace = True)
```

```python
df.drop('song_formatted', axis =1 , inplace = True)
df.to_csv('filtered_final.csv', index = False)
```
